```python
corr = df.corr()
corr
```

```python
corr_class = corr['Class'].sort_values(ascending=False)
corr_class
```

```python
plt.figure(figsize=(10,6))
corr['Class'].sort_values(ascending=False).plot(kind='bar')
plt.title('Correlação de cada feature coma a variável alvo (Class)')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(corr, cmap='coolwarm_r')
plt.title('Mapa de calor da correlação entre as features')
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.plot(df['Time'], alpha=0.7)
plt.title("Distribuição geral das transações ao longo do tempo")
plt.xlabel("Index")
plt.ylabel("Time (segundos)")
plt.show()
```

```python
fraudes = df[df['Class'] == 1]

plt.figure(figsize=(12,8))
plt.scatter(fraudes.index, fraudes['Time'], s=10, alpha=0.6)
plt.title("Horário das transações fraudulentas")
plt.xlabel("Index da transação")
plt.ylabel("Time (segundos)")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.hist(df['Time'], bins=50)
plt.title("Histograma de todas as transações por tempo")
plt.xlabel("Time (segundos)")
plt.ylabel("Frequência")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.hist(fraudes['Time'], bins=50, color='red')
plt.title("Histograma apenas de fraudes")
plt.xlabel("Time (segundos)")
plt.ylabel("Frequência")
plt.show()
```

tende a ter mais fraudes nas transacoes que demoram menos tempos e de valores mais abaixos de 5k
